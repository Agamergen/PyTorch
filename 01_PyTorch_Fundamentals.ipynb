### There are two main rules that performing matrix multiplication needs to satisfy:

1. The **INNER DIMENSIONS** must match

* '(3, 2) @ (3, 2)' WON'T work
* '(2, 3) @ (3, 2)' WILL work
* '(3, 2) @ (2, 3)' WILL work

2. The resulting matrix has the shape of the **OUTER DIMENSIONS**:

* '(2, 3) @ (3, 2)' -> '(2, 2)' WILL work **OUTER**
* '(3, 2) @ (2, 3)' -> '(3, 3)' WILL work **OUTER**

```python
# Inner dimension 7, 7
torch.matmul(torch.rand(4, 7), torch.rand(7, 4))
```

```python
# Outer dimension 3, 3
torch.matmul(torch.rand(3, 5), torch.rand(5, 3))
```

One of the most common errors in deep learning: shape errors.

For matrix multiplication - http://matrixmultiplication.xyz/

```python
# Shapes for matrix multiplication

tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
torch.mm(tensor_A, tensor_B)  # torch.mm is the same as torch.matmul
```

```python
tensor_A.shape, tensor_B.shape
```

### It should be [3, 2], [2, 3]

1. So to fix our tensor shape issues, we can manipulate the shape of one of our tensors using a **TRANSPOSE**
2. A **TRANSPOSE** switches the axes or dimensions of a given tensor


```python
# tensor_B.T - T stands for TRANSPOSE

tensor_B.T,  tensor_B.T.shape
```

```python
tensor_B, tensor_B.shape
```

```python
torch.matmul(tensor_A, tensor_B.T).shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed

print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

In Python, the `squeeze()` function from the numpy library is used to remove single-dimensional entries from the shape of an array. The parameter you pass to `squeeze()` is the axis (or axes) that you want to squeeze.

However, it's important to note that `squeeze()` can only remove axes with a length of 1 (i.e., dimensions that have only one element). If a dimension has more than one element, `squeeze()` cannot remove it.

So, if you have an array with shape (3, 4, 2) and you want to squeeze the last dimension (which has 2 elements), using `squeeze(2)` would not change the array because the last dimension is not a single-element dimension.

If you have an array with shape (3, 1, 2) and you use `squeeze(1)`, it will remove the second dimension which is a single-element dimension, resulting in a new array with shape (3, 2).

Yes, you got it!

In the shape (3, 1, 2), you can think of the **3 as layers** (like floors in a building), the **1 as a row** (like a single apartment on a floor), and the **2 as columns** (like rooms in the apartment).

When you squeeze out the '1', it's like removing the concept of 'apartment', and you're left with just floors (layers) and rooms (columns). So, your new shape (3, 2) represents 3 layers, each with 2 rooms. Well done! 

Actually, the method would be `unsqueeze(1)`. Here's why:

In our building analogy, if you want to add the concept of 'apartments' back into your building, you would need to add it in the right place. Right now, you have 3 layers (floors) and 2 rooms (columns). If you want each floor to have multiple apartments, you would add the 'apartment' dimension back in between the 'floors' and 'rooms'.

In array terms, this means adding a dimension in the second position (since we start counting from 0). So you would use `unsqueeze(1)` to get a new shape of (3, 1, 2), which represents 3 layers (floors), each with 1 apartment, and each apartment with 2 rooms.

### REPRODUCIBILITY (trying to take random out of random)

In short how neural network learns:

'start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again ...'

To reduce the randomness in nerual network and PyTorch comes the concept of a **RANDOM SEED**
Essentially what the random seed does is 'flavour' the randomness.

```python
import torch

# Create two random tensors

random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
import torch

# Let's make random but reproducible tensors

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)

```
